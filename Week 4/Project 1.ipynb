First, I created a .csv file titled Project_1_Data.csv. I will import import pandas and numpy. I will use pandas to read the .csv file.


```python
import pandas as pd
import numpy as np
arrival_delays = pd.read_csv("C:/Users/pamel/Dropbox/CUNY SPS/IS 362 - Data Acquisition and Mgmt/Week 4/Project_1_Data.csv")
arrival_delays
```

Next, I'm making a copy of the arrival delays data to preserve the original.

```python
flight_data = arrival_delays.copy()
```

Next, I will create a column showing the total amount of flights

```python
flight_data["Total_flights"] = flight_data.sum(axis = 1)
flight_data
```

Next, I will create a column showing the percentage of flights that are delayed


```python
flight_data['Delays_percentage'] = flight_data["Delayed"]/flight_data["Total_flights"]
flight_data
```

Next, I will sort the data by percentage of delays from lowest to highest


```python
flight_data.sort_values(by = "Delays_percentage")
```

Finally, I will pivot the dataframe for an easy comparison of the airlines, showing the delayed flights percentage for each destination.


```python
flight_data_pivot = flight_data.pivot(columns = "Airline", index = "Destination", values = ["Delays_percentage"])
flight_data_pivot
```
