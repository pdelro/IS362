The following is my analysis of the student results.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
student_results = pd.read_csv("https://raw.githubusercontent.com/pdelro/IS362/main/Week%206/student_results.csv")
student_results
```

I will split the sex and age column into two separate columns

```python
student_results[['sex', 'age']] = student_results['sex and age'].str.split('_', expand = True)
student_results
```

I will now create a column for the average test scores for each student

```python
student_results['average'] = student_results[['term 1', 'term 2', 'term 3']].mean(axis=1)
student_results
```

I will display how many students took each test split by sex

```python
plt.figure(figsize = (5,5))
test_count = sns.countplot('test number', hue='sex', data= student_results)
```

I will show the results for each student for each test by term

```python
plt.figure(figsize=(10,5))
term_1 = sns.barplot('name', 'term 1', hue='test number', data= student_results)
```

```python
plt.figure(figsize=(10,5))
term_2 = sns.barplot('name', 'term 2', hue='test number', data= student_results)
```

```python
plt.figure(figsize=(10,5))
term_3 = sns.barplot('name', 'term 3', hue='test number', data= student_results)
```

Finally, I will show the average of each test by sex

```python
plt.figure(figsize=(10,5))
term_3 = sns.barplot('test number', 'average', hue='sex', data= student_results)
```
