```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#scrape uswnt 2021 data from the url https://www.ussoccer.com/uswnt-stats/2021
uswnt_html = pd.read_html('https://www.ussoccer.com/uswnt-stats/2021', header=0)
#only want to scrape the data up to the last player so that data is uniform with the .csv used later. goal keepers often do not have any goals, I will replace NaN with 0
uswnt_2021 = uswnt_html[0][:37].fillna(0)
uswnt_2021
```

```python
#insert 2021 column so that data is uniform with the .csv file for 2015-2020 data
uswnt_2021.insert(0, 'Year', '2021')
uswnt_2021
```

```python
#display top 10 goal scorers from the 2021 roster
top_ten_goal_scorers_women = uswnt_2021.sort_values(by = ['Goals'], ascending=False).head(10)
top_ten_goal_scorers_women
```

```python
#display top ten goal scorers in barplot
plt.figure(figsize=(12,5))
total_goals_by_player = sns.barplot(x = 'Goals', y = 'Name', data = top_ten_goal_scorers_women)
```

```python
#Is there a correlation between caps (player's appearance in a game) and goals?
plt.figure(figsize = (15,5))
#choosing uswnt_2021 data since caps and goals are measured career measurements, not just specific to the year
sns.scatterplot(data=uswnt_2021, x= 'Caps', y = 'Goals')
```

There seems to be no correlation between caps and goals. My hypothesis is that the position of the player has a positive correlation with goals.

```python
plt.figure(figsize=(10,5))
goals_by_position = sns.barplot('Pos.', 'Goals', data=uswnt_2021)
```

This barplot shows that most goals were scored by forwards, then midfielders, and then finally defenders which is logical.

```python
#I would like to find stats by player for the years 2015-2021
#create dataframe from .csv file
uswnt_2015_2020 = pd.read_csv('https://raw.githubusercontent.com/pdelro/IS362/main/Final%20Project/uswnt_cleaned.csv')
uswnt_2015_2020
```

```python
#combine data into one dataframe
uswnt = pd.concat([uswnt_2015_2020, uswnt_2021], axis=0)
uswnt
```

```python
player_stats = uswnt.copy()
player_stats
```

```python
by_player = player_stats.groupby('Name', as_index=False).sum()
by_player
```

```python
#since caps and goals are career stats, there sum is not meaningful so I will exclude these columns
by_player.drop(['Caps','Goals'], axis=1, inplace=True)
by_player
```

```python
#I want to display the top ten players who had the highest assists from the years 2015-2021
top_ten_assist_makers = by_player.sort_values(by = ['A'], ascending=False).head(10)
top_ten_assist_makers
```

```python
#display top ten assist makers in barplot
plt.figure(figsize=(12,5))
total_assists_by_player = sns.barplot(x = 'A', y = 'Name', data = top_ten_assist_makers)
```

```python
by_position = player_stats.groupby(['Year','Pos.'], as_index=False).sum()
by_position.drop(['Caps','Goals'], axis=1, inplace=True)
by_position
```

```python
#display goals by year by position
plt.figure(figsize=(15,5))
term_1 = sns.barplot('Year', 'G', hue='Pos.', data= by_position)
```

```python
#display assists by year by position
plt.figure(figsize=(15,5))
term_1 = sns.barplot('Year', 'A', hue='Pos.', data= by_position)
```

```python
#scrape 2021 usmnmt data from the url https://www.ussoccer.com/usmnt-stats/2021
usmnt_html = pd.read_html('https://www.ussoccer.com/usmnt-stats/2021', header=0)
#only want to scrape the data up to the last player so that data is uniform with the .csv used later. goal keepers often do not have any goals, I will replace NaN with 0
usmnt_2021 = usmnt_html[0][:57].fillna(0)
usmnt_2021
```

```python
#for a quick and simple comparison, display top 10 goal scorers from the 2021 roster of the USMNT
top_ten_goal_scorers_men = usmnt_2021.sort_values(by = ['Career Goals'], ascending=False).head(10)
top_ten_goal_scorers_men
```

```python
top_ten_goal_scorers_men.rename(columns={'Player':'Name','Career Goals': 'Goals'}, inplace=True)
top_ten_goal_scorers_men
```

```python
#combine top ten goals scorers for both men and women into one dataframe
df1 = pd.DataFrame(top_ten_goal_scorers_women, columns=['Name','Goals'])
df1.insert(0, 'Gender', 'F')
df1
```

```python
df2 = pd.DataFrame(top_ten_goal_scorers_men, columns=['Name','Goals'])
df2.insert(0, 'Gender', 'M')
df2
```

```python
top_ten_goal_scorers = pd.concat([df1, df2], axis=0)
top_ten_goal_scorers
```

```python
top_ten_goal_scorers = pd.concat([df1, df2], axis=0)
top_ten_goal_scorers
```

```python
#display total goals by gender
plt.figure(figsize=(10,5))
goals_by_gender = sns.barplot(x = 'Gender', y = 'Goals', data=top_ten_goal_scorers, estimator=sum)
```
